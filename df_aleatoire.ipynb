```python
import pandas as pd
import random

# Listes de valeurs possibles
intitule_poste = ["Data Ingenieur", "Data Engineer", "Data Analyst", "Data Analyste", "Data Scientist"]
departements = ["Ain", "Aisne", "Allier", "Alpes-de-Haute-Provence", "Hauts-de-Seine", "Seine-Saint-Denis", "Val-de-Marne", "Val-d'Oise"]
niveau_experience = ["Junior", "IntermÃ©diaire", "Senior"]
secteur_d_activite = ["Tech", "Finance", "Industrie", "SantÃ©", "Marketing", "Autre"]
taille_entreprise = ["Startup", "PME", "Grande entreprise"]
type_de_contrat = ["CDI", "CDD", "Freelance", "Stage", "Alternance"]
teletravail = ["Complet", "Partiel", "Occasionnel", "Aucun"]

def generer_nom_entreprise():
    prefixes = ["Tech", "Data", "AI", "Cloud", "Innov", "Cyber", "Smart"]
    suffixes = ["Corp", "Solutions", "Industries", "Labs", "Consulting", "Systems", "Services"]
    return random.choice(prefixes) + random.choice(suffixes) + " " + str(random.randint(1, 99))

def generer_description(row):
    return (f"ğŸ“Œ **Poste** : {row['intitule_poste']} ({row['niveau_experience']})\n"
            f"ğŸ¢ **Entreprise** : {row['nom_entreprise']} ({row['taille_entreprise']})\n"
            f"ğŸ“ **Localisation** : {row['departement']}\n"
            f"ğŸ“Š **Secteur** : {row['secteur_d_activite']}\n"
            f"ğŸ“ **Type de contrat** : {row['type_de_contrat']}\n"
            f"ğŸ’° **Salaire estimÃ©** : {row['salaire']} â‚¬\n"
            f"ğŸ’» **TÃ©lÃ©travail** : {row['teletravail']}\n\n"
            f"ğŸ” **Description** : Nous recherchons un {row['intitule_poste']} motivÃ© pour rejoindre notre Ã©quipe dynamique "
            f"chez {row['nom_entreprise']}. Vous serez amenÃ©(e) Ã  travailler sur des projets innovants dans le domaine "
            f"{row['secteur_d_activite'].lower()}. Vous aurez l'opportunitÃ© de collaborer avec des experts et "
            f"de dÃ©velopper vos compÃ©tences dans un environnement stimulant. \n\n"
            f"ğŸ¯ **Missions** : \n"
            f"- Analyser et interprÃ©ter les donnÃ©es pour aider Ã  la prise de dÃ©cision\n"
            f"- DÃ©velopper des modÃ¨les prÃ©dictifs et des visualisations avancÃ©es\n"
            f"- Collaborer avec les Ã©quipes mÃ©tiers pour comprendre leurs besoins\n"
            f"- Automatiser et optimiser les processus analytiques\n\n"
            f"âœ… **Profil recherchÃ©** : \n"
            f"- ExpÃ©rience : {row['niveau_experience']} en data science ou analyse de donnÃ©es\n"
            f"- CompÃ©tences : Python, SQL, Tableau, Machine Learning\n"
            f"- Esprit dâ€™analyse, rigueur et bon relationnel\n\n"
            f"ğŸ“© **Postulez dÃ¨s maintenant !**")

# GÃ©nÃ©rer 200 lignes de donnÃ©es fictives
data = {
    "intitule_poste": [random.choice(intitule_poste) for _ in range(200)],
    "departement": [random.choice(departements) for _ in range(200)],
    "niveau_experience": [random.choice(niveau_experience) for _ in range(200)],
    "secteur_d_activite": [random.choice(secteur_d_activite) for _ in range(200)],
    "taille_entreprise": [random.choice(taille_entreprise) for _ in range(200)],
    "type_de_contrat": [random.choice(type_de_contrat) for _ in range(200)],
    "teletravail": [random.choice(teletravail) for _ in range(200)],
    "salaire": [random.randint(30000, 100000) for _ in range(200)],
    "nom_entreprise": [generer_nom_entreprise() for _ in range(200)]
}

# CrÃ©ation du DataFrame
df = pd.DataFrame(data)

# Ajout des colonnes annonce et lien
df["annonce"] = df.apply(generer_description, axis=1)
df["lien"] = ["https://job-platform.com/offre-" + str(i) for i in range(1, 201)]

# Affichage des premiÃ¨res lignes
df.head()
```
